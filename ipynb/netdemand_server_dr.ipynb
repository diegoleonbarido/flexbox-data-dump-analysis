```python
import pandas as pd
from datetime import datetime, timedelta, date
from niuera_nica import psql_niuera_real_time
from niuera_nica import psql_niuera_cndc_hourly
from niuera_nica import psql_niuera_cndc_20sec as psql_20sec
metadata = psql_niuera_real_time.get_metadata()
table_dict = psql_niuera_real_time.setup_tables(metadata)
import matplotlib.pyplot as plt
%matplotlib inline  


#### Actual CNDC Data POSDESPACHO

metadata = psql_niuera_cndc_hourly.get_metadata()
table_dict = psql_niuera_cndc_hourly.setup_tables(metadata)

column_names = table_dict['posdespacho_table'].columns.keys()
df_hourly_post = pd.DataFrame(table_dict['posdespacho_table'].select().\
     order_by(table_dict['posdespacho_table'].c.datetime.asc()).execute().fetchall(),columns=column_names)
df_hourly_post.index = df_hourly_post['datetime']
df_hourly_post['date'] = df_hourly_post.index.date
df_hourly_post['hour'] = df_hourly_post.index.hour

df_hourly_post['gen_wind'] = df_hourly_post['AMY1'] + df_hourly_post['AMY2']+ df_hourly_post['PBP'] + df_hourly_post['EOL'] + df_hourly_post['ABR']
df_hourly_post['load_inter_total'] = df_hourly_post['LNI-L9040'] + df_hourly_post['SND-L9090']+ df_hourly_post['AMY-L9030'] + df_hourly_post['TCPI-L9150']
df_hourly_post['net_demand'] = df_hourly_post['Demanda'] - df_hourly_post['gen_wind'] # + df_hourly_post['load_inter_total']

#(df_real['demand_ni']/10).plot()
#(df_real['demand_ni']/10)[datetime(2016,6,20):datetime(2016,6,21,23)].plot()
#df_hourly['Demanda'].plot()


#### Actual CNDC Data PREDESPACHO

column_names = table_dict['predespacho_table'].columns.keys()
df_hourly = pd.DataFrame(table_dict['predespacho_table'].select().\
                         order_by(table_dict['predespacho_table'].c.datetime.asc()).execute().fetchall(),columns=column_names)
df_hourly.index = df_hourly['datetime']
df_hourly['date'] = df_hourly.index.date
df_hourly['hour'] = df_hourly.index.hour

df_hourly['gen_wind'] = df_hourly['AMY1'] + df_hourly['AMY2']+ df_hourly['PBP'] + df_hourly['EOL'] + df_hourly['ABR']
df_hourly['load_inter_total'] = df_hourly['LNI-L9040'] + df_hourly['SND-L9090']+ df_hourly['AMY-L9030'] + df_hourly['TCPI-L9150']
df_hourly['net_demand'] = df_hourly['Demanda'] - df_hourly['gen_wind'] + df_hourly['load_inter_total']

df_hourly_resample = df_hourly['net_demand'].resample('20S').mean().interpolate(method='time')


#### 5 Second Demanda Data

metadata = psql_niuera_real_time.get_metadata()
table_dict = psql_niuera_real_time.setup_tables(metadata)

column_names = table_dict['eor_5sec_table'].columns.keys()
df_real = pd.DataFrame(table_dict['eor_5sec_table'].select().\
     order_by(table_dict['eor_5sec_table'].c.datetime.asc()).execute().fetchall(),columns=column_names)
df_real.index = df_real['datetime']
df_real['date'] = df_real.index.date
df_real['hour'] = df_real.index.hour


####  20 Second Wind Data

metadata = psql_20sec.get_metadata()
table = psql_20sec.setup_tables(metadata)

column_names = table_dict['cndc_20sec_table'].columns.keys()
twenty_sec = pd.DataFrame(table_dict['cndc_20sec_table'].select().\
        #where(cast(table_dict['predespacho_table'].c.datetime,Date)>date_now).\
        order_by(table_dict['cndc_20sec_table'].c.datetime.asc()).execute().fetchall(),columns=column_names)
twenty_sec.index = twenty_sec['datetime']
twenty_sec['date'] = twenty_sec.index.date
twenty_sec['hour'] = twenty_sec.index.hour

wind_plants = ['gen_EOL','gen_ABR','gen_AMY','gen_PBP']
wind_plant_limits = [45,40,63,40]

#Dropping Wind Values that Might be Outliers
twenty_sec.loc[twenty_sec['gen_EOL'] > 45*10, 'gen_EOL'] = None
twenty_sec.loc[twenty_sec['gen_ABR'] > 40*10, 'gen_ABR'] = None
twenty_sec.loc[twenty_sec['gen_AMY'] > 63*10, 'gen_AMY'] = None
twenty_sec.loc[twenty_sec['gen_PBP'] > 40*10, 'gen_PBP'] = None

twenty_sec['gen_wind'] = (twenty_sec['gen_EOL'] + twenty_sec['gen_ABR'] + twenty_sec['gen_AMY'] + twenty_sec['gen_PBP'])/10

#Dropping Demand Values that Might be Outliers
twenty_sec['load_total'] =  (twenty_sec['load_BZN'] +  twenty_sec['load_CHG'] + twenty_sec['load_acoyapa'] + twenty_sec['load_altamira'] + twenty_sec['load_amerrisque'] + twenty_sec['load_asososca']+ twenty_sec['load_asturias'] + twenty_sec['load_batahola'] + twenty_sec['load_bluefields']+twenty_sec['load_boaco'] +  twenty_sec['load_chinandega'] + twenty_sec['load_corinto'] + twenty_sec['load_corocito'] + twenty_sec['load_diriamba']+twenty_sec['load_el_mojon'] + twenty_sec['load_el_periodista'] + twenty_sec['load_el_tuma']+twenty_sec['load_el_viejo'] + twenty_sec['load_enacal'] + twenty_sec['load_esteli']+ twenty_sec['load_granada']+twenty_sec['load_la_esperanza'] + twenty_sec['load_la_gateada'] + twenty_sec['load_las_banderas']+ twenty_sec['load_leon_1']+twenty_sec['load_leon_2'] + twenty_sec['load_los_brasiles'] + twenty_sec['load_malpaisillo']+ twenty_sec['load_managua']+twenty_sec['load_masatepe'] + twenty_sec['load_matagalpa'] + twenty_sec['load_matiguas_mulukuku_siuna']+ twenty_sec['load_nadaime']+twenty_sec['load_oriental'] + twenty_sec['load_portezuelo'] + twenty_sec['load_punta_huete']+ twenty_sec['load_rivas']+twenty_sec['load_san_benito'] + twenty_sec['load_san_miguelito'] + twenty_sec['load_san_ramon']+ twenty_sec['load_sandino']+twenty_sec['load_santa_clara'] + twenty_sec['load_sebaco'] + twenty_sec['load_ticuantepe_2']+ twenty_sec['load_tipitapa'] +  twenty_sec['load_yalaguina'])/10
twenty_sec.loc[twenty_sec['load_total'] > 850, 'load_total'] = None

twenty_sec['load_inter_total'] = (twenty_sec['inter_LNI-L9040'] + twenty_sec['inter_SND-L9090'] + twenty_sec['inter_AMY-L9030']+ twenty_sec['inter_TCPI-L9150'])/10
twenty_sec.loc[twenty_sec['load_inter_total'] < -100, 'load_inter_total'] = None
twenty_sec.loc[twenty_sec['load_inter_total'] > 100, 'load_inter_total'] = None

#Net Demand
twenty_sec['net_demand'] = twenty_sec['load_total'] - twenty_sec['gen_wind'] + twenty_sec['load_inter_total'] 


####### Merge Between the 20 Second Data (Wind) and the 5 Second Data (Demand)

twenty_sec_subset = twenty_sec[['datetime','date','hour','gen_wind','load_inter_total']].resample('20S').mean().interpolate(method='time')
df_real_subset = df_real[['datetime','date','hour','demand_ni']].resample('20S').mean().interpolate(method='time')
df_netdemand_20sec = pd.merge(twenty_sec_subset,df_real_subset,left_index=True, right_index=True)[['hour_x','gen_wind','demand_ni','load_inter_total']]
df_netdemand_20sec['net_demand_20sec'] = (df_netdemand_20sec['demand_ni']/10) - df_netdemand_20sec['gen_wind'] - df_netdemand_20sec['load_inter_total']

df_netdemand_20sec['date'] = df_netdemand_20sec.index.date
df_netdemand_20sec['hour'] = df_netdemand_20sec.index.hour
```

```python
# 20 Second Data

start_sub_date = datetime(2016,7,1,13,36,20)
day_dfs = pd.DataFrame()

for i in range(0,184):

    time_min = i * 5 
    
    day_slice0 = all_data_df[start_sub_date + timedelta(minutes=time_min) : start_sub_date + timedelta(minutes=time_min + 10)]
    day_slice1 = all_data_df[start_sub_date + timedelta(minutes= time_min + 10) : start_sub_date + timedelta(minutes= time_min + 20)]
    hourly_pred_slice = all_data_df[start_sub_date + timedelta(minutes=time_min): start_sub_date + timedelta(minutes= time_min + 20)]
        
    if (day_slice0.shape[0] != 0) and (day_slice1.shape[0] != 0):
                
        ## Net Demand
        slope_demand_diff1 = (day_slice1['net_demand_20sec'].tail(1)[0] - day_slice1['net_demand_20sec'].head(1)[0])/(20*30)
        
        # Mean during subset time periods
        demand_mean0 = day_slice0['net_demand_20sec'].mean()
        demand_mean1 = day_slice1['net_demand_20sec'].mean()
        predicted_mean = hourly_pred_slice['net_demand'].mean()
        
        day_slice1['event_netdemand'] = [None] * len(day_slice1['hour'])
        if (slope_demand_diff1 > 0.005) and (demand_mean1 > demand_mean0) and (demand_mean1 > predicted_mean):
            day_slice1['event_netdemand'] = 1
        else:
            day_slice1['event_netdemand'] = 0
            
        ## Appending the slice    
        day_slice1['event_id'] = i   
        day_dfs.append(day_slice1)    
        
        if i == 0:
            day_slice_df = day_slice1
        else:
            day_slice_df = pd.concat([day_slice_df,day_slice1],axis=0)
```

```python
plt.plot(day_slice_df.index,day_slice_df['net_demand'],'.')
#plt.plot(df_hourly_post_subset.index,df_hourly_post_subset['Demanda'])
plt.plot(day_slice_df.index,day_slice_df['net_demand_20sec'])
plt.plot(day_slice_df.index,day_slice_df['event_netdemand']*300)

plt.legend()
```
