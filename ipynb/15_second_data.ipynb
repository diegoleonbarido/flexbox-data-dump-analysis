If everything goes well, the data should be loaded and this notebook should show values. There are a lot of values so it may take awhile to load into a dataframe.

```python
metadata = psql.get_metadata()
table_ref = psql.setup_tables(metadata)
```

```python
values = table_ref.select()
result = values.execute()
df = pd.DataFrame(result.fetchall(),columns=result.keys())
df = df.drop_duplicates(subset='datetime').dropna()
df= df.set_index('datetime')
df = df/10 #We store all values in the database as integers, but they are really decimals (135 is 13.5)
#There is one generator that I called gen_Unknown because it is a number on the map isn't near
#any of the words, but I wanted to collect it anyway so we could eventually identify it.
df = df.sort_index()
#We had some data before April 1 but some of the generators were mixed up in the wrong columns, so this is all
#of the data since I fixed the way that the numbers are associated with the labels.
```
