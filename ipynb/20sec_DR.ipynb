```python
# 20 Second Data

# Save date as a good example for paper start_sub_date = datetime(2016,6,12,0,0)


start_sub_date = datetime(2016,6,12,0,0)
day_dfs = pd.DataFrame()

for i in range(0,184):

    time_min = i * 10 
    
    day_slice0 = day_subset[start_sub_date + timedelta(minutes=time_min) : start_sub_date + timedelta(minutes=time_min + 10)]
    day_slice1 = day_subset[start_sub_date + timedelta(minutes= time_min + 10) : start_sub_date + timedelta(minutes= time_min + 20)]
    
    if (day_slice0.shape[0] != 0) and (day_slice1.shape[0] != 0):
    
        # Extracting data for the 
        day_pred_slice = day_subset_pred[day_subset_pred['hour'] == day_slice1['hour'][0]]
        
        
        ## Wind
        
        # Differences at the beginning of every period
        wind_diff0 = day_slice0['gen_wind'].tail(1)[0] - day_slice0['gen_wind'].head(1)[0] 
        wind_diff1 = day_slice1['gen_wind'].tail(1)[0] - day_slice1['gen_wind'].head(1)[0] 
        
        # Mean during time periods
        wind_mean0 = day_slice0['gen_wind'].mean()
        wind_mean1 = day_slice1['gen_wind'].mean()
        
        # Difference between wind and actual wind
        day_slice1['wind_tty_pred_diff'] = ((wind_mean1 - day_pred_slice['gen_wind'][0])/day_pred_slice['gen_wind'][0])*100
        
        day_slice1['event_wind'] = [None] * len(day_slice1['hour'])
        if ((wind_diff1 < 0) or (wind_diff1 >=0 and wind_diff1 <1)) and (wind_mean1 < wind_mean0) and (wind_mean1 < day_pred_slice['gen_wind'][0]):
            day_slice1['event_wind'] = 1
        else:
            day_slice1['event_wind'] = 0
            
        ## Demand
        
        slope_demand_diff1 = (day_slice1['load_total'].tail(1)[0] - day_slice1['load_total'].head(1)[0])/600
        # Mean during time periods
        demand_mean0 = day_slice0['load_total'].mean()
        demand_mean1 = day_slice1['load_total'].mean()
        
        day_slice1['event_demand'] = [None] * len(day_slice1['hour'])
        if (slope_demand_diff1 > 0.005) and (demand_mean1 > demand_mean0):
            day_slice1['event_demand'] = 1
        else:
            day_slice1['event_demand'] = 0
        
        day_slice1['twentysec_dr_event'] = day_slice1['event_demand'] + day_slice1['event_wind'] 
            
        ## Appending the slice    
        
        day_slice1['event_id'] = i   
        day_dfs.append(day_slice1)    
            
        if i == 0:
            day_slice_df = day_slice1
        else:
            day_slice_df = pd.concat([day_slice_df,day_slice1],axis=0)
        
```

```python
import pandas as pd
import os
from pandas.tools.plotting import scatter_matrix
from datetime import date, datetime, timedelta
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib as mpl
%matplotlib inline
import seaborn as sns
sns.set_style("white")
import statsmodels.api as sm
from sklearn import linear_model
from scipy.stats import norm,ttest_ind
from scipy import stats
import copy
import numpy as np
import os.path
from ggplot import *
from matplotlib.cm import cool
import re
from niuera_nica import psql_niuera_cndc_hourly as psql_hour
from sqlalchemy import cast,Date,text
from niuera_nica import psql_niuera_cndc_20sec as psql_20sec

##### Variables to change at the top of the file (NOTE: Change the variables below afterwards, they were being 
    # used for testing). Change the two variables below whenever you are testing it outside the function and comment out the other ones

#date_now = datetime(2016,6,12,0,0)
#date_now_sec = datetime(2016,6,12,0,0)

def dr_20_sec(this_date): 

    #####  Importing Data from the Pre-Despacho Table for the Hour
    
    metadata = psql_hour.get_metadata()
    table_dict = psql_hour.setup_tables(metadata)
    column_names = table_dict['predespacho_table'].columns.keys()
    
    # Grabbing only the top of the hour
    date_now = this_date.replace(minute=0,second=0,microsecond=0)
    
    pred_frame = pd.DataFrame(table_dict['predespacho_table'].select().\
            where(table_dict['predespacho_table'].c.datetime == date_now).\
            order_by(table_dict['predespacho_table'].c.datetime.asc()).execute().fetchall(),columns=column_names)
    
    pred_frame.index = pd.to_datetime(pred_frame['datetime'])
    
    pred_frame['date'] = pred_frame.index.date
    pred_frame['hour'] = pred_frame.index.hour
    
    pred_frame['gen_wind'] = pred_frame['AMY1'] + pred_frame['AMY2']+ pred_frame['PBP'] + pred_frame['EOL'] + pred_frame['ABR']
    pred_frame['load_inter_total'] = pred_frame['LNI-L9040'] + pred_frame['SND-L9090']+ pred_frame['AMY-L9030'] + pred_frame['TCPI-L9150']
    pred_frame['net_demand'] = pred_frame['Demanda'] - pred_frame['gen_wind'] + pred_frame['load_inter_total']
    
    
    
    #####  Importing data from SQL Table
    
    #Generation and Prices Data Frame
    metadata = psql_20sec.get_metadata()
    table = psql_20sec.setup_tables(metadata)
    column_names = table.columns.keys()
    
    # Grabbing only the top of the hour
    date_now_sec = this_date
    
    twenty_sec = pd.DataFrame(table.select().\
            where(cast(table.c.datetime,Date)>date_now_sec - timedelta(minutes=20)).\
            order_by(table.c.datetime.asc()).execute().fetchall(),columns=column_names)
    
    twenty_sec.index = pd.to_datetime(twenty_sec['datetime'])
    
    twenty_sec['date'] = twenty_sec.index.date
    twenty_sec['hour'] = twenty_sec.index.hour
    
    
    
    ######   Editing the Wind Data and Creating the 20 Second Database
    
    wind_plants = ['gen_EOL','gen_ABR','gen_AMY','gen_PBP']
    wind_plant_limits = [45,40,63,40]
    
    #Dropping Wind Values that Might be Outliers
    twenty_sec.loc[twenty_sec['gen_EOL'] > 45*10, 'gen_EOL'] = None
    twenty_sec.loc[twenty_sec['gen_ABR'] > 40*10, 'gen_ABR'] = None
    twenty_sec.loc[twenty_sec['gen_AMY'] > 63*10, 'gen_AMY'] = None
    twenty_sec.loc[twenty_sec['gen_PBP'] > 40*10, 'gen_PBP'] = None
    
    twenty_sec['gen_wind'] = (twenty_sec['gen_EOL'] + twenty_sec['gen_ABR'] + twenty_sec['gen_AMY'] + twenty_sec['gen_PBP'])/10
    
    #Dropping Demand Values that Might be Outliers
    twenty_sec['load_total'] =  (twenty_sec['load_BZN'] +  twenty_sec['load_CHG'] + twenty_sec['load_acoyapa'] + twenty_sec['load_altamira'] + twenty_sec['load_amerrisque'] + twenty_sec['load_asososca']+ twenty_sec['load_asturias'] + twenty_sec['load_batahola'] + twenty_sec['load_bluefields']+twenty_sec['load_boaco'] +  twenty_sec['load_chinandega'] + twenty_sec['load_corinto'] + twenty_sec['load_corocito'] + twenty_sec['load_diriamba']+twenty_sec['load_el_mojon'] + twenty_sec['load_el_periodista'] + twenty_sec['load_el_tuma']+twenty_sec['load_el_viejo'] + twenty_sec['load_enacal'] + twenty_sec['load_esteli']+ twenty_sec['load_granada']+twenty_sec['load_la_esperanza'] + twenty_sec['load_la_gateada'] + twenty_sec['load_las_banderas']+ twenty_sec['load_leon_1']+twenty_sec['load_leon_2'] + twenty_sec['load_los_brasiles'] + twenty_sec['load_malpaisillo']+ twenty_sec['load_managua']+twenty_sec['load_masatepe'] + twenty_sec['load_matagalpa'] + twenty_sec['load_matiguas_mulukuku_siuna']+ twenty_sec['load_nadaime']+twenty_sec['load_oriental'] + twenty_sec['load_portezuelo'] + twenty_sec['load_punta_huete']+ twenty_sec['load_rivas']+twenty_sec['load_san_benito'] + twenty_sec['load_san_miguelito'] + twenty_sec['load_san_ramon']+ twenty_sec['load_sandino']+twenty_sec['load_santa_clara'] + twenty_sec['load_sebaco'] + twenty_sec['load_ticuantepe_2']+ twenty_sec['load_tipitapa'] +  twenty_sec['load_yalaguina'])/10
    twenty_sec.loc[twenty_sec['load_total'] > 850, 'load_total'] = None
    
    twenty_sec['load_inter_total'] = (twenty_sec['inter_LNI-L9040'] + twenty_sec['inter_SND-L9090'] + twenty_sec['inter_AMY-L9030']+ twenty_sec['inter_TCPI-L9150'])/10
    twenty_sec.loc[twenty_sec['load_inter_total'] < -100, 'load_inter_total'] = None
    twenty_sec.loc[twenty_sec['load_inter_total'] > 100, 'load_inter_total'] = None
    
    #Net Demand
    twenty_sec['net_demand'] = twenty_sec['load_total'] - twenty_sec['gen_wind'] + twenty_sec['load_inter_total'] 
    
    #Day Subset 
    day_subset = twenty_sec[['hour','load_total','gen_wind','net_demand','load_inter_total']]
    
    #Slicing the Data
    day_slice0 = day_subset[this_date : this_date + timedelta(minutes=10)]
    day_slice1 = day_subset[this_date : this_date + timedelta(minutes=20)]    
    
    
    ##### Beginning of Comparison
    
    if (day_slice0.shape[0] != 0) and (day_slice1.shape[0] != 0):
        
         ## Wind
            
        # Differences at the beginning of every period
        wind_diff0 = day_slice0['gen_wind'].tail(1)[0] - day_slice0['gen_wind'].head(1)[0] 
        wind_diff1 = day_slice1['gen_wind'].tail(1)[0] - day_slice1['gen_wind'].head(1)[0] 
        
        # Mean during time periods
        wind_mean0 = day_slice0['gen_wind'].mean()
        wind_mean1 = day_slice1['gen_wind'].mean()
        
        # Difference between wind and actual wind
        day_slice1['wind_tty_pred_diff'] = ((wind_mean1 - pred_frame['gen_wind'][0])/pred_frame['gen_wind'][0])*100
        
        day_slice1['event_wind'] = [None] * len(day_slice1['hour'])
        if ((wind_diff1 < 0) or (wind_diff1 >=0 and wind_diff1 <1)) and (wind_mean1 < wind_mean0) and (wind_mean1 < pred_frame['gen_wind'][0]):
            day_slice1['event_wind'] = 1
        else:
            day_slice1['event_wind'] = 0
            
            
        ## Demand
        slope_demand_diff1 = (day_slice1['load_total'].tail(1)[0] - day_slice1['load_total'].head(1)[0])/600
        # Mean during time periods
        demand_mean0 = day_slice0['load_total'].mean()
        demand_mean1 = day_slice1['load_total'].mean()
        
        ## Creating the Day Slice
        day_slice1['event_demand'] = [None] * len(day_slice1['hour'])
         
        if (slope_demand_diff1 > 0.005) and (demand_mean1 > demand_mean0):
            day_slice1['event_demand'] = 1
        else:
            day_slice1['event_demand'] = 0
        
        day_slice1['twentysec_dr_event'] = day_slice1['event_demand'] + day_slice1['event_wind'] 
        
    if sum(day_slice1['twentysec_dr_event']) >0:
        return(1)
    else:
        return(0)
```
