```python
def read_wind_data(source,date_select,resample_choice):
    
    wind_pred_dic = {}
    
    
    ##### 1. Hourly Post Despacho Data
    
    # Server Connection
    if source == "server":
        engine = create_engine('postgresql://niuera_analyzer:analysis@cooljoule.javirosa.com/cndc_hourly_db')
        metadata = MetaData(bind=engine)
    else:
        metadata = psql_hour.get_metadata()
    
    table_dict = psql_hour.setup_tables(metadata)
    column_names = table_dict['posdespacho_table'].columns.keys()
    
    post_frame = pd.DataFrame(table_dict['posdespacho_table'].select().\
                              where(table_dict['posdespacho_table'].c.datetime>=date_select).\
                              order_by(table_dict['posdespacho_table'].c.datetime.asc()).execute().fetchall(),columns=column_names)
    post_frame.index = post_frame['datetime']
    post_frame['date'] = post_frame.index.date
    post_frame['hour'] = post_frame.index.hour
    
    post_frame['gen_wind'] = post_frame['AMY1'] + post_frame['AMY2']+ post_frame['PBP'] + post_frame['EOL'] + post_frame['ABR']
    post_frame['load_inter_total'] = post_frame['LNI-L9040'] + post_frame['SND-L9090']+ post_frame['AMY-L9030'] + post_frame['TCPI-L9150']
    post_frame['net_demand'] = post_frame['Demanda'] - post_frame['gen_wind'] + post_frame['load_inter_total']

    ##### 2. Hourly Prediction Data
    
    # Server Connection
    if source == "server":
        engine = create_engine('postgresql://niuera_analyzer:analysis@cooljoule.javirosa.com/cndc_hourly_db')
        metadata = MetaData(bind=engine)
    else:
        metadata = psql_hour.get_metadata()
    
    table_dict = psql_hour.setup_tables(metadata)
    column_names = table_dict['predespacho_table'].columns.keys()
    
    pred_frame = pd.DataFrame(table_dict['predespacho_table'].select().\
                              where(table_dict['predespacho_table'].c.datetime>=date_select).\
                              order_by(table_dict['predespacho_table'].c.datetime.asc()).execute().fetchall(),columns=column_names)
    pred_frame.index = pred_frame['datetime']
    pred_frame['date'] = pred_frame.index.date
    pred_frame['hour'] = pred_frame.index.hour
    
    pred_frame['gen_wind'] = pred_frame['AMY1'] + pred_frame['AMY2']+ pred_frame['PBP'] + pred_frame['EOL'] + pred_frame['ABR']
    pred_frame['load_inter_total'] = pred_frame['LNI-L9040'] + pred_frame['SND-L9090']+ pred_frame['AMY-L9030'] + pred_frame['TCPI-L9150']
    pred_frame['net_demand'] = pred_frame['Demanda'] - pred_frame['gen_wind'] + pred_frame['load_inter_total']
    
    
    ##### 3. Real Time 20 Second and 5 Second Data
    
    # Server Connection
    
    if source == "server":
        engine = create_engine('postgresql://niuera_analyzer:analysis@cooljoule.javirosa.com/niuera_real_time_db')
        metadata = MetaData(bind=engine)
    else: 
        metadata = psql_niuera_real_time.get_metadata()
    
    
    table_dict = psql_niuera_real_time.setup_tables(metadata)
    
    # Extracting 20 Second Data Tables
    column_names = table_dict['cndc_20sec_table'].columns.keys()
    twenty_sec = pd.DataFrame(table_dict['cndc_20sec_table'].select().\
                              where(table_dict['cndc_20sec_table'].c.datetime>=date_select).\
                              order_by(table_dict['cndc_20sec_table'].c.datetime.asc()).execute().fetchall(),columns=column_names)
    twenty_sec.index = twenty_sec['datetime']
    twenty_sec['date'] = twenty_sec.index.date
    twenty_sec['hour'] = twenty_sec.index.hour
    
    wind_plants = ['gen_EOL','gen_ABR','gen_AMY','gen_PBP']
    wind_plant_limits = [45,40,63,40]
    
    #Dropping Wind Values that Might be Outliers
    twenty_sec.loc[twenty_sec['gen_EOL'] > 45*10, 'gen_EOL'] = None
    twenty_sec.loc[twenty_sec['gen_ABR'] > 40*10, 'gen_ABR'] = None
    twenty_sec.loc[twenty_sec['gen_AMY'] > 63*10, 'gen_AMY'] = None
    twenty_sec.loc[twenty_sec['gen_PBP'] > 40*10, 'gen_PBP'] = None
    
    twenty_sec['gen_wind'] = (twenty_sec['gen_EOL'] + twenty_sec['gen_ABR'] + twenty_sec['gen_AMY'] + twenty_sec['gen_PBP'])/10
    
    #Dropping Demand Values that Might be Outliers
    twenty_sec['load_total'] =  (twenty_sec['load_BZN'] +  twenty_sec['load_CHG'] + twenty_sec['load_acoyapa'] + twenty_sec['load_altamira'] + twenty_sec['load_amerrisque'] + twenty_sec['load_asososca']+ twenty_sec['load_asturias'] + twenty_sec['load_batahola'] + twenty_sec['load_bluefields']+twenty_sec['load_boaco'] +  twenty_sec['load_chinandega'] + twenty_sec['load_corinto'] + twenty_sec['load_corocito'] + twenty_sec['load_diriamba']+twenty_sec['load_el_mojon'] + twenty_sec['load_el_periodista'] + twenty_sec['load_el_tuma']+twenty_sec['load_el_viejo'] + twenty_sec['load_enacal'] + twenty_sec['load_esteli']+ twenty_sec['load_granada']+twenty_sec['load_la_esperanza'] + twenty_sec['load_la_gateada'] + twenty_sec['load_las_banderas']+ twenty_sec['load_leon_1']+twenty_sec['load_leon_2'] + twenty_sec['load_los_brasiles'] + twenty_sec['load_malpaisillo']+ twenty_sec['load_managua']+twenty_sec['load_masatepe'] + twenty_sec['load_matagalpa'] + twenty_sec['load_matiguas_mulukuku_siuna']+ twenty_sec['load_nadaime']+twenty_sec['load_oriental'] + twenty_sec['load_portezuelo'] + twenty_sec['load_punta_huete']+ twenty_sec['load_rivas']+twenty_sec['load_san_benito'] + twenty_sec['load_san_miguelito'] + twenty_sec['load_san_ramon']+ twenty_sec['load_sandino']+twenty_sec['load_santa_clara'] + twenty_sec['load_sebaco'] + twenty_sec['load_ticuantepe_2']+ twenty_sec['load_tipitapa'] +  twenty_sec['load_yalaguina'])/10
    twenty_sec.loc[twenty_sec['load_total'] > 850, 'load_total'] = None
    
    twenty_sec['load_inter_total'] = (twenty_sec['inter_LNI-L9040'] + twenty_sec['inter_SND-L9090'] + twenty_sec['inter_AMY-L9030']+ twenty_sec['inter_TCPI-L9150'])/10
    twenty_sec.loc[twenty_sec['load_inter_total'] < -100, 'load_inter_total'] = None
    twenty_sec.loc[twenty_sec['load_inter_total'] > 100, 'load_inter_total'] = None
    
    #Net Demand
    twenty_sec['net_demand'] = twenty_sec['load_total'] - twenty_sec['gen_wind'] #+ twenty_sec['load_inter_total'] 
    
    twenty_sec_resample = twenty_sec.resample(resample_choice).mean().interpolate(method='time')
    
    
    # Extracting 5 Second Data Tables
    column_names = table_dict['eor_5sec_table'].columns.keys()
    five_sec = pd.DataFrame(table_dict['eor_5sec_table'].select().\
                      where(table_dict['eor_5sec_table'].c.datetime>=date_select).\
                      order_by(table_dict['eor_5sec_table'].c.datetime.asc()).execute().fetchall(),columns=column_names)
    five_sec.index = five_sec['datetime']
    five_sec['date'] = five_sec.index.date
    five_sec['hour'] = five_sec.index.hour
    
    five_sec['demand_5sec'] = five_sec['demand_ni']/10
    five_sec['cr_ni_5sec'] = five_sec['cr_ni']/100
    five_sec['ni_cr_5sec'] = five_sec['ni_cr']/100
    five_sec['ho_ni_5sec'] = five_sec['ho_ni']/100
    five_sec['ni_ho_5sec'] = five_sec['ni_ho']/100
    
    five_sec_resample = five_sec.resample(resample_choice).mean().interpolate(method='time')
    
    
    ##### 4.  Bringing the DR Events
    
    # Server Connection
    if source == "server":
        engine = create_engine('postgresql://niuera_analyzer:analysis@cooljoule.javirosa.com/niuera_control_db')
        metadata = MetaData(bind=engine)
    else:
        metadata = psql_niuera_control.get_metadata()
    
    table_dict = psql_niuera_control.setup_tables(metadata)
    
    # Net Load DR 
    column_names = table_dict['net_load_dr_table'].columns.keys()
    net_load_dr = pd.DataFrame(table_dict['net_load_dr_table'].select().\
                              where(table_dict['net_load_dr_table'].c.datetime>=date_select).\
                              order_by(table_dict['net_load_dr_table'].c.datetime.asc()).execute().fetchall(),columns=column_names)
    net_load_dr.index = net_load_dr['datetime']
    net_load_dr['date'] = net_load_dr.index.date
    net_load_dr['hour'] = net_load_dr.index.hour
    
    # Peak Shifting DR Table
    column_names = table_dict['peak_shifting_dr_table'].columns.keys()
    peak_shifting_dr = pd.DataFrame(table_dict['peak_shifting_dr_table'].select().\
                              where(table_dict['peak_shifting_dr_table'].c.datetime>=date_select).\
                              order_by(table_dict['peak_shifting_dr_table'].c.datetime.asc()).execute().fetchall(),columns=column_names)
    
    peak_shifting_dr.index = peak_shifting_dr['datetime']
    peak_shifting_dr['date'] = peak_shifting_dr.index.date
    peak_shifting_dr['hour'] = peak_shifting_dr.index.hour
    
    wind_pred_dic['post_frame'] = post_frame
    wind_pred_dic['pred_frame'] = pred_frame
    wind_pred_dic['twenty_sec'] = twenty_sec_resample
    wind_pred_dic['five_sec'] = five_sec_resample
    wind_pred_dic['net_load_dr'] = net_load_dr
    wind_pred_dic['peak_shifting_dr'] = peak_shifting_dr
    
    return (wind_pred_dic);
```

## Compare the Models

```python
def compare_models(data,resample_choice,variable,ma,ar):
    
    data_resample = data.resample(resample_choice).mean()

    df1 = data_resample[np.isfinite(data_resample[variable])][variable]
    arma_mod = sm.tsa.ARMA(df1, (ma,ar)).fit()
    print("Model with MA " + str(ma) + " and AR " + str(ar) +  " has")
    print()
    print(arma_mod.params)
    print()
    print("and AIC: "  + str(arma_mod.aic) + ", BIC: " + str(arma_mod.bic) + ", HQIC: " + str(arma_mod.hqic))
    print()
    print('The Durbin Watson statistic is: ' + str(sm.stats.durbin_watson(arma_mod.resid.values)) + 'values closer to 2 are sign of a stationary series') 
    
    #Meaning of the autocorrelated lags
    
    r,q,p = sm.tsa.acf(arma_mod.resid.values.squeeze(), qstat=True)
    data = np.c_[range(1,41), r[1:], q, p]
    table = pd.DataFrame(data, columns=['lag', "AC", "Q", "Prob(>Q)"])
    print()
    print('The statistical significance of the autocorrelated lags is as follows. The null is tha the lag is non-zero')
    print(table.set_index('lag'))
    
    # Plotting Residuals
    
    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(111)
    ax = arma_mod.resid.plot(ax=ax);
    plt.title('Residuals')
    
    plt.figure()
    
    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(111)
    ax = plt.hist(arma_mod.resid, bins=np.arange(min(arma_mod.resid), max(arma_mod.resid) + 1, 1))
    plt.title('Histogram of Residuals')
    
    plt.figure()
    
    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(111)
    fig = qqplot(arma_mod.resid, line='q', ax=ax, fit=True)
    
    plt.figure()
    
   
```

```python
def pac_plots(data,resample_choice,lag_choice):
    
    data_resampled = data.resample(resample_choice).mean() 
    
    fig = plt.figure(figsize=(12,8))
    plt.plot(data_resampled.index,data_resampled['net_demand'])
    plt.plot(data_resampled.index,data_resampled['gen_wind'])
    
    plt.figure()

    df1 = data_resampled[np.isfinite(data_resampled['net_demand'])]
    
    fig = plt.figure(figsize=(12,8))
    ax1 = fig.add_subplot(211)
    fig = sm.graphics.tsa.plot_acf(df1['net_demand'].values.squeeze(), lags=lag_choice, ax=ax1)
    ax2 = fig.add_subplot(212)
    fig = sm.graphics.tsa.plot_pacf(df1['net_demand'], lags=lag_choice, ax=ax2)
    
    plt.figure()
    
    df2 = data_resampled[np.isfinite(data_resampled['gen_wind'])]
    
    fig = plt.figure(figsize=(12,8))
    ax1 = fig.add_subplot(211)
    fig = sm.graphics.tsa.plot_acf(df2['gen_wind'].values.squeeze(), lags=lag_choice, ax=ax1)
    ax2 = fig.add_subplot(212)
    fig = sm.graphics.tsa.plot_pacf(df2['gen_wind'], lags=lag_choice, ax=ax2)
```

```python
def iterative_arma(data,variable,pred_resample_choice,train_resample_choice,training_date_begin,training_date_end,pred_date_begin,pred_date_end,time_delta,ma,ar):
    
    # Data for training
    df_training = data[variable].resample(train_resample_choice).mean()
    
    # Data to be predicted
    df_resample_choice = data[variable].resample(pred_resample_choice).mean()
    predict_set_resample = df_resample_choice[((df_resample_choice.index >= pred_date_begin) & (df_resample_choice.index < pred_date_end))]
    
    predicted_vals = []
    time_delta_count = -time_delta
    
    for i in range(0,predict_set_resample.index.shape[0]): # For each 5 minute interval in the range we want to predict
        
        time_delta_count += time_delta
        train_set_dynamic = df_training[(df_training.index >= training_date_begin) & (df_training.index < training_date_end + timedelta(minutes=time_delta_count))]
        arma_mod = sm.tsa.ARMA(train_set_dynamic, (ma,ar)).fit()
        predict_df = arma_mod.predict(str(train_set_dynamic.index[-1] + timedelta(minutes=1)),str(train_set_dynamic.index[-1] + timedelta(minutes=20)), dynamic=False)
        #predict_df = arma_mod.predict(str(train_set_dynamic.index[-1] + timedelta(minutes=0)),str(train_set_dynamic.index[-1] + timedelta(minutes=20)), dynamic=False)
        
        
        predicted_vals.append(predict_df)
        
        
    # Putting the data together
    for i in range(0,len(predicted_vals)):
        
        if i == 0:
            pd_data_frame = pd.DataFrame(predicted_vals[i])
            pd_data_frame['datetime'] = pd_data_frame.index
            pd_data_frame.rename(columns={0:variable}, inplace=True)
        else:
            pd_data_concat = pd.DataFrame(predicted_vals[i])
            pd_data_concat['datetime'] = pd_data_concat.index
            pd_data_concat.rename(columns={0:variable}, inplace=True)
            
            pd_data_frame = pd.concat([pd_data_frame, pd_data_concat], axis=0)
    
    df_predicted = pd_data_frame.groupby(pd_data_frame['datetime']).mean()
    
    actual = data[(data.index >= pred_date_begin) & (data.index < pred_date_end)]
                 
    plt.plot(actual.index,actual['gen_wind'],color='blue')
    plt.plot(df_predicted.index,df_predicted['gen_wind'],'.',color='red')
    
    plt.figure() 
```

```python
def iterative_arma_dr(data,variable,pred_resample_choice,train_resample_choice,training_date_begin,training_date_end,pred_date_begin,pred_date_end,time_delta,ma,ar,predict_ahead):

    # Data for training
    df_training = data[variable].resample(train_resample_choice).mean()
    
    # Data to be predicted
    df_resample_choice = data[variable].resample(pred_resample_choice).mean()
    predict_set_resample = df_resample_choice[((df_resample_choice.index >= pred_date_begin) & (df_resample_choice.index < pred_date_end))]
    
    predicted_vals = []
    time_delta_count = -time_delta
    
    
    #####
    
    wind_mean1_list = []
    wind_mean0_list = []
    wind_diff0mw_list = []
    wind_diff1mw_list = []
    wind_diff1_list = []
    wind_diff0_list = []
    wind_events = []
    time_events = []
    predict_df_list = []
    
    for i,val in enumerate(predict_set_resample.index): # For each 5 minute interval in the range we want to predict
        
        time_delta_count += time_delta
        train_set_dynamic = df_training[(df_training.index >= training_date_begin) & (df_training.index < training_date_end + timedelta(minutes=time_delta_count))]
        arma_mod = sm.tsa.ARMA(train_set_dynamic, (ma,ar)).fit()
        
        if train_resample_choice == '5T':
            predict_df = arma_mod.predict(str(train_set_dynamic.index[-1] + timedelta(minutes=0)),str(train_set_dynamic.index[-1] + timedelta(minutes=predict_ahead)), dynamic=False)
            predicted_vals.append(predict_df)
            
        else:
            predict_df = arma_mod.predict(str(train_set_dynamic.index[-1] + timedelta(minutes=1)),str(train_set_dynamic.index[-1] + timedelta(minutes=predict_ahead)), dynamic=False)
            predicted_vals.append(predict_df)
        
        ####
        
        # Grab the last fifeteen minutes
        last_twenty  = pd.DataFrame(train_set_dynamic[(train_set_dynamic.index >= (train_set_dynamic.index[-1] - timedelta(minutes=15)))\
                                     & (train_set_dynamic.index <= train_set_dynamic.index[-1])])
        last_twenty['date'] = last_twenty.index.date
        last_twenty['hour'] = last_twenty.index.hour
        
        # Grab the last thirty minutes
        last_thirty  = pd.DataFrame(train_set_dynamic[(train_set_dynamic.index >= (train_set_dynamic.index[-1] - timedelta(minutes=30)))\
                                     & (train_set_dynamic.index <= train_set_dynamic.index[-1] - timedelta(minutes=15))])
        last_thirty['date'] = last_thirty.index.date
        last_thirty['hour'] = last_thirty.index.hour
        
        
        day_slice0 = last_twenty
        day_slice1 = pd.DataFrame(predict_df)
        day_slice1.rename(columns={0:'gen_wind'}, inplace=True)
        day_slice2 = last_thirty
        day_slice2.rename(columns={0:'gen_wind'}, inplace=True)
        
        
         ##### Beginning of Comparison
        if (day_slice0.shape[0] != 0) and (day_slice1.shape[0] != 0):
            
             ## Wind
                
            # Differences at the beginning of every period
            wind_diff0 = day_slice0['gen_wind'].tail(1)[0] - day_slice0['gen_wind'].head(1)[0] 
            wind_diff1 = day_slice1['gen_wind'].tail(1)[0] - day_slice1['gen_wind'].head(1)[0] 
            wind_diff1_mw = ((day_slice1['gen_wind'].tail(1)[0] - day_slice1['gen_wind'].head(1)[0]) / (day_slice1['gen_wind'].head(1)[0]))
            wind_diff0_mw = ((day_slice0['gen_wind'].tail(1)[0] - day_slice0['gen_wind'].head(1)[0]) / (day_slice0['gen_wind'].head(1)[0]))
            
            # Mean during time periods
            wind_mean0 = day_slice0['gen_wind'].mean()
            wind_mean1 = day_slice1['gen_wind'].mean()
            wind_mean2 = day_slice2['gen_wind'].mean()
            
            # Difference between wind and actual wind
            day_slice1['wind_tty_pred_diff'] = ((wind_mean1 - pred_frame['gen_wind'][0])/pred_frame['gen_wind'][0])*100
            
            
            wind_mean1_list.append(wind_mean1)
            wind_mean0_list.append(wind_mean0)
            wind_diff0mw_list.append(wind_diff0_mw)
            wind_diff1mw_list.append(wind_diff1_mw)
            wind_diff1_list.append(wind_diff1)
            wind_diff0_list.append(wind_diff0)
            
            #Note: we began taking iterative means, then we move to an ARMA process (since our signal is stationary)
            #if (((wind_diff0 > 0) and (wind_diff1 <0)) and wind_diff1_mw < -0.055 and wind_diff0_mw > 0.04) and np.mean([wind_mean0,wind_mean1]) >= 50:
            #if ((wind_diff1 <0) and (wind_diff1_mw < -0.01) and (np.mean([wind_mean0,wind_mean1]) >= 50)):
            #if ((wind_diff1 <0) and (wind_diff1_mw < -0.01)):
            
            # Need to have these two if else statements, else the algorithm won't run
            
            #if (i <=1):
            #    if (((wind_mean1 - wind_mean0_list[-1])/wind_mean0_list[-1]) < -0.05):
            #        event_wind = 1
            #        wind_events.append(event_wind)
            #        time_events.append(val)
            #    else:
            #        event_wind = 0
            #        wind_events.append(event_wind)
            #        time_events.append(val)
            #else:
            #    if (((wind_mean1 - wind_mean0_list[-3])/wind_mean0_list[-3]) < -0.05):
            #        event_wind = 1
            #        wind_events.append(event_wind)
            #        time_events.append(val)
            #    else:
            #        event_wind = 0
            #        wind_events.append(event_wind)
            #        time_events.append(val)
            
            # Change between wind_mean0 if you want to measure sudden drops in wind, the other one would be a little delayed
            
            if (((wind_mean1 - wind_mean0)/wind_mean0) < -0.05):
                event_wind = 1
                wind_events.append(event_wind)
                time_events.append(val)
            else:
                event_wind = 0
                wind_events.append(event_wind)
                time_events.append(val)
                
                    
    # Predictions
    for i in range(0,len(predicted_vals)):
        
        if i == 0:
            pd_data_frame = pd.DataFrame(predicted_vals[i])
            pd_data_frame['datetime'] = pd_data_frame.index
            pd_data_frame.rename(columns={0:variable}, inplace=True)
        else:
            pd_data_concat = pd.DataFrame(predicted_vals[i])
            pd_data_concat['datetime'] = pd_data_concat.index
            pd_data_concat.rename(columns={0:variable}, inplace=True)
            
            pd_data_frame = pd.concat([pd_data_frame, pd_data_concat], axis=0)
    
    df_predicted = pd_data_frame.groupby(pd_data_frame['datetime']).mean()
    
    actual = data[(data.index >= pred_date_begin) & (data.index < pred_date_end)]
                 
    plt.plot(actual.index,actual['gen_wind'],color='blue')
    plt.xticks(rotation=90)
    plt.plot(df_predicted.index,df_predicted['gen_wind'],'.',color='red')
              
                    
    # Wind Events 
    pd_wind = pd.DataFrame(wind_events)
    pd_wind.rename(columns={0:'signal_wind'}, inplace=True)
    pd_datetimes = pd.DataFrame(time_events)
    pd_datetimes.rename(columns={0:'datetimes_wind'}, inplace=True)
    
    pd_wind = pd.concat([pd_wind, pd_datetimes], axis=1)
    pd_wind.index = pd_wind['datetimes_wind']
    
    
    # Demand & Wind: 5 & 20 Second Data
    twenty_sec_sub = twenty_sec_resample[(twenty_sec_resample.index >= pred_date_begin) & (twenty_sec_resample.index < pred_date_end)]
    plt.plot(twenty_sec_sub.index,twenty_sec_sub['gen_wind'])
    
    # Net Load DR Events
    net_load_sub = pd_wind[(pd_wind['datetimes_wind'] > pred_date_begin) & (pd_wind['datetimes_wind'] < pred_date_end)]
    net_load_dr_events = (net_load_sub[net_load_sub['signal_wind'] == 1]).index.unique()
    
    for event in net_load_dr_events:
        plt.axvspan(event,event + timedelta(minutes=20), alpha=0.1, color='blue')
        
    net_load_df = (net_load_sub[net_load_sub['signal_wind'] == 1])
    net_load_df['hour'] = net_load_df.index.hour
    net_load_df['date'] = net_load_df.index.date
    
    hour = net_load_df.groupby(['date','hour']).first()
    hour.index = hour['datetimes_wind']
    hourly_events = hour.index 
    
    for event in hourly_events:
        plt.axvspan(event,event + timedelta(minutes=20), alpha=0.6, color='red')
        
        
    plt.figure()   
```
