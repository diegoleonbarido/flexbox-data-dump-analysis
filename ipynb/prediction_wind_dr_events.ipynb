```python
def pac_plots(data,resample_choice,lag_choice):
    
    data_resampled = data.resample(resample_choice).mean() 
    
    fig = plt.figure(figsize=(12,8))
    plt.plot(data_resampled.index,data_resampled['net_demand'])
    plt.plot(data_resampled.index,data_resampled['gen_wind'])
    
    plt.figure()

    df1 = data_resampled[np.isfinite(data_resampled['net_demand'])]
    
    fig = plt.figure(figsize=(12,8))
    ax1 = fig.add_subplot(211)
    fig = sm.graphics.tsa.plot_acf(df1['net_demand'].values.squeeze(), lags=lag_choice, ax=ax1)
    ax2 = fig.add_subplot(212)
    fig = sm.graphics.tsa.plot_pacf(df1['net_demand'], lags=lag_choice, ax=ax2)
    
    plt.figure()
    
    df2 = data_resampled[np.isfinite(data_resampled['gen_wind'])]
    
    fig = plt.figure(figsize=(12,8))
    ax1 = fig.add_subplot(211)
    fig = sm.graphics.tsa.plot_acf(df2['gen_wind'].values.squeeze(), lags=lag_choice, ax=ax1)
    ax2 = fig.add_subplot(212)
    fig = sm.graphics.tsa.plot_pacf(df2['gen_wind'], lags=lag_choice, ax=ax2)
```

```python
def iterative_arma(data,variable,pred_resample_choice,train_resample_choice,training_date_begin,training_date_end,pred_date_begin,pred_date_end,time_delta,ma,ar):
    
    # Data for training
    df_training = data[variable].resample(train_resample_choice).mean()
    
    # Data to be predicted
    df_resample_choice = data[variable].resample(pred_resample_choice).mean()
    predict_set_resample = df_resample_choice[((df_resample_choice.index >= pred_date_begin) & (df_resample_choice.index < pred_date_end))]
    
    predicted_vals = []
    time_delta_count = -time_delta
    
    for i in range(0,predict_set_resample.index.shape[0]): # For each 5 minute interval in the range we want to predict
        
        time_delta_count += time_delta
        train_set_dynamic = df_training[(df_training.index >= training_date_begin) & (df_training.index < training_date_end + timedelta(minutes=time_delta_count))]
        arma_mod = sm.tsa.ARMA(train_set_dynamic, (ma,ar)).fit()
        predict_df = arma_mod.predict(str(train_set_dynamic.index[-1] + timedelta(minutes=1)),str(train_set_dynamic.index[-1] + timedelta(minutes=20)), dynamic=False)
        
        predicted_vals.append(predict_df)
        
        
    # Putting the data together
    for i in range(0,len(predicted_vals)):
        
        if i == 0:
            pd_data_frame = pd.DataFrame(predicted_vals[i])
            pd_data_frame['datetime'] = pd_data_frame.index
            pd_data_frame.rename(columns={0:variable}, inplace=True)
        else:
            pd_data_concat = pd.DataFrame(predicted_vals[i])
            pd_data_concat['datetime'] = pd_data_concat.index
            pd_data_concat.rename(columns={0:variable}, inplace=True)
            
            pd_data_frame = pd.concat([pd_data_frame, pd_data_concat], axis=0)
    
    df_predicted = pd_data_frame.groupby(pd_data_frame['datetime']).mean()
    
    actual = data[(data.index >= pred_date_begin) & (data.index < pred_date_end)]
                 
    plt.plot(actual.index,actual,color='blue')
    plt.plot(df_predicted.index,df_predicted,color='red')
    
    plt.figure() 
```

```python
def iterative_arma_dr(data,variable,pred_resample_choice,train_resample_choice,training_date_begin,training_date_end,pred_date_begin,pred_date_end,time_delta,ma,ar,predict_ahead):

    # Data for training
    df_training = data[variable].resample(train_resample_choice).mean()
    
    # Data to be predicted
    df_resample_choice = data[variable].resample(pred_resample_choice).mean()
    predict_set_resample = df_resample_choice[((df_resample_choice.index >= pred_date_begin) & (df_resample_choice.index < pred_date_end))]
    
    predicted_vals = []
    time_delta_count = -time_delta
    
    
    #####
    
    wind_mean1_list = []
    wind_mean0_list = []
    wind_diff0mw_list = []
    wind_diff1mw_list = []
    wind_diff1_list = []
    wind_diff0_list = []
    wind_events = []
    time_events = []
    predict_df_list = []
    
    for i,val in enumerate(predict_set_resample.index): # For each 5 minute interval in the range we want to predict
        
        time_delta_count += time_delta
        train_set_dynamic = df_training[(df_training.index >= training_date_begin) & (df_training.index < training_date_end + timedelta(minutes=time_delta_count))]
        arma_mod = sm.tsa.ARMA(train_set_dynamic, (ma,ar)).fit()
        predict_df = arma_mod.predict(str(train_set_dynamic.index[-1] + timedelta(minutes=1)),str(train_set_dynamic.index[-1] + timedelta(minutes=predict_ahead)), dynamic=False)
        predicted_vals.append(predict_df)
        
        ####
        
        # Grab the last ten minutes
        last_twenty  = pd.DataFrame(train_set_dynamic[(train_set_dynamic.index >= (train_set_dynamic.index[-1] - timedelta(minutes=10)))\
                                     & (train_set_dynamic.index <= train_set_dynamic.index[-1])])
        last_twenty['date'] = last_twenty.index.date
        last_twenty['hour'] = last_twenty.index.hour
        
        day_slice0 = last_twenty
        day_slice1 = pd.DataFrame(predict_df)
        day_slice1.rename(columns={0:'gen_wind'}, inplace=True)
        
        
         ##### Beginning of Comparison
        if (day_slice0.shape[0] != 0) and (day_slice1.shape[0] != 0):
            
             ## Wind
                
            # Differences at the beginning of every period
            wind_diff0 = day_slice0['gen_wind'].tail(1)[0] - day_slice0['gen_wind'].head(1)[0] 
            wind_diff1 = day_slice1['gen_wind'].tail(1)[0] - day_slice1['gen_wind'].head(1)[0] 
            wind_diff1_mw = ((day_slice1['gen_wind'].tail(1)[0] - day_slice1['gen_wind'].head(1)[0]) / (day_slice1['gen_wind'].head(1)[0]))
            wind_diff0_mw = ((day_slice0['gen_wind'].tail(1)[0] - day_slice0['gen_wind'].head(1)[0]) / (day_slice0['gen_wind'].head(1)[0]))
            
            # Mean during time periods
            wind_mean0 = day_slice0['gen_wind'].mean()
            wind_mean1 = day_slice1['gen_wind'].mean()
            
            # Difference between wind and actual wind
            day_slice1['wind_tty_pred_diff'] = ((wind_mean1 - pred_frame['gen_wind'][0])/pred_frame['gen_wind'][0])*100
            
            
            wind_mean1_list.append(wind_mean1)
            wind_mean0_list.append(wind_mean0)
            wind_diff0mw_list.append(wind_diff0_mw)
            wind_diff1mw_list.append(wind_diff1_mw)
            wind_diff1_list.append(wind_diff1)
            wind_diff0_list.append(wind_diff0)
            
            #if (((wind_diff0 > 0) and (wind_diff1 <0)) and wind_diff1_mw < -0.055 and wind_diff0_mw > 0.04) and np.mean([wind_mean0,wind_mean1]) >= 50:
            #if ((wind_diff1 <0) and (wind_diff1_mw < -0.01) and (np.mean([wind_mean0,wind_mean1]) >= 50)):
            #if ((wind_diff1 <0) and (wind_diff1_mw < -0.01)):
            
            # Need to have these two if else statements, else the algorithm won't run
            
            if (i <=1):
                if (((wind_mean1 - wind_mean0_list[-1])/wind_mean0_list[-1]) < -0.10):
                    event_wind = 1
                    wind_events.append(event_wind)
                    time_events.append(val)
                else:
                    event_wind = 0
                    wind_events.append(event_wind)
                    time_events.append(val)
            else:
                if (((wind_mean1 - wind_mean0_list[-2])/wind_mean0_list[-2]) < -0.10):
                    event_wind = 1
                    wind_events.append(event_wind)
                    time_events.append(val)
                else:
                    event_wind = 0
                    wind_events.append(event_wind)
                    time_events.append(val)
    
                    
    # Wind
    pd_wind = pd.DataFrame(wind_events)
    pd_wind.rename(columns={0:'signal_wind'}, inplace=True)
    pd_datetimes = pd.DataFrame(time_events)
    pd_datetimes.rename(columns={0:'datetimes_wind'}, inplace=True)
    
    pd_wind = pd.concat([pd_wind, pd_datetimes], axis=1)
    pd_wind.index = pd_wind['datetimes_wind']
    
    
    # Demand & Wind: 5 & 20 Second Data
    twenty_sec_sub = twenty_sec_resample[(twenty_sec_resample.index >= pred_date_begin) & (twenty_sec_resample.index < pred_date_end)]
    plt.plot(twenty_sec_sub.index,twenty_sec_sub['gen_wind'])
    
    # Net Load DR Events
    net_load_sub = pd_wind[(pd_wind['datetimes_wind'] > pred_date_begin) & (pd_wind['datetimes_wind'] < pred_date_end)]
    net_load_dr_events = (net_load_sub[net_load_sub['signal_wind'] == 1]).index.unique()
    
    for event in net_load_dr_events:
        plt.axvspan(event,event + timedelta(minutes=20), alpha=0.3, color='blue')
            
```
