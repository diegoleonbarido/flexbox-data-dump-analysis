```python
def create_monotonically_increasing_energy_vals_mfilocal(table_name,column_list,flxbx,server_boolean=True,threshold_boolean=True):
    '''
    table_name: mfi, or zwave, or wambient
    flxb: used on the server when we're brining in data with many ids
    server_boolean: this is True if we are running it on the server with many flxbx IDs
    server_boolean: this is True if we are running it on the server with many flxbx IDs
    '''
    column_names = table_name.columns.keys()
    df = pd.DataFrame(table_name.select().order_by(table_name.c.datetime.asc()).execute().fetchall(),columns=column_names)
    df.index = pd.to_datetime(df['datetime'])
    
    temporary_df = copy.deepcopy(df)
     
    for name in column_list:    
        for val in df[name].loc[df[name].diff()<-1].index:
            temporary_df[name][val:]=\
            temporary_df[name][val:]-df[name].diff()[val]
                    
    return temporary_df
```

```python
column_names_fridge = table_dict['mfi_table'].columns.keys()
column_names_inside = table_dict['inside_table'].columns.keys()
column_names_ambient = table_dict['ambient_table'].columns.keys()

# The next two lines are SQL Alchemy code, ordering each house energy data by date and then executing 
fridge = pd.DataFrame(table_dict['mfi_table'].select().order_by(table_dict['mfi_table'].c.datetime.asc()).execute()\
                      .fetchall(),columns=column_names_fridge).set_index('datetime')
inside = pd.DataFrame(table_dict['inside_table'].select().order_by(table_dict['inside_table'].c.datetime.asc()).\
                      execute().fetchall(),columns=column_names_inside).set_index('datetime')
ambient = pd.DataFrame(table_dict['ambient_table'].select().order_by(table_dict['ambient_table'].c.datetime.asc()).\
                       execute().fetchall(),columns=column_names_ambient).set_index('datetime')

# Resample to Second
fridge = fridge.resample('60S').mean()
inside = inside.resample('60S').mean()
ambient = ambient.resample('60S').mean()

# Merge
state_df = inside.join(fridge,how='left',lsuffix='_x') 
state_df = state_df.join(ambient,how='left',lsuffix='_x') 
```

```python
state_df.index = state_df.index - timedelta(hours=6)
```
