```python
#All Values in MW Capacity
#Fossil Fuels
df_cap = {}
df_cap['CEN'] = 57
df_cap['PCG1'] = 19.2
df_cap['PCG2'] = 19.2
df_cap['PCG3'] = 19.2
df_cap['PCG4'] = 19.2
df_cap['PCG5'] = 18.6
df_cap['PCG6'] = 12.9
df_cap['PCG7'] = 38
df_cap['PCG8'] = 25.3
df_cap['PCG9'] = 45.3
df_cap['PHC1'] = 15
df_cap['PHC2'] = 45
df_cap['EEC20'] = 18.5
df_cap['EEC'] = 50
df_cap['TPC'] = 50.9
df_cap['PLB1'] = 23
df_cap['PLB2'] = 35
df_cap['PMG3'] = 42
df_cap['PMG4'] = 5.5
df_cap['PMG5'] = 5.5
df_cap['PNI1'] = 50
df_cap['PNI2'] = 50
df_cap['GSR'] = 5.4

#Wind
df_cap['ABR'] = 39.6
df_cap['AMY1'] = 39.9
df_cap['AMY2'] = 23.1
df_cap['EOL'] = 44.4
df_cap ['PBP'] = 39.6

#Hydro
df_cap['PCA1'] = 25
df_cap['PCA2'] = 25
df_cap['PCF1'] = 25
df_cap['PCF2'] = 25
df_cap['HPA1'] = 6.5
df_cap['HPA2'] = 6.5
df_cap['PHL1'] = 8.5
df_cap['PHL2'] = 8.5

#Geothermal
df_cap['PEN1 y 2'] = 5
df_cap['PEN3'] = 36
df_cap['PEN4'] = 36
df_cap['PMT1'] = 30
df_cap['PMT2'] = 30
df_cap['PMT3'] = 6.4
df_cap['PCHN'] = 15

#Biomass
df_cap['MTR'] = 35
df_cap['NSL'] = 30

def get_cfs(start,end):
    cap_fact = []
    cap_dict = {}
    keys = []
    for val in df_actual.keys():
        if val in df_cap:
            cf = float(sum(df_actual[val][start:end])/(df_cap[val]*len(df_actual[start:end])))
            cap_dict[val] = cf
            cap_fact.append(cf)
            keys.append(val)
    df_cf = pd.DataFrame([keys,cap_fact]).T
    df_cf.columns = ['Plant','CF']
    df_cf.sort_values('CF',ascending=False)
    return cap_dict
```
